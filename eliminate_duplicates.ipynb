```python
import pickle
from tqdm import tqdm_notebook as tqdm
```

```python
fileloc = "data/dataset_documents_facts_non_facts20200427.pickle"
with open(fileloc, "rb") as file:
    data = pickle.load(file)
```

```python
trimmed_data = []
```

```python
freq = {}
for datum in tqdm(data):
    if " ".join(datum["facts"]) in freq:
        freq[" ".join(datum["facts"])] += 1
    else:
        freq[" ".join(datum["facts"])] = 1
```

```python
for i, k in enumerate(freq):
    print(i, freq[k])
```

```python
freqs = {}
for datum in tqdm(data):
    facts = " ".join(datum["facts"])
    if facts in freqs:
        freqs[facts] += 1
    else:
        freqs[facts] = 1
        trimmed_data.append(datum)
    
```

```python
with open("data/dataset_documents_facts_non_facts20200518.pickle", "wb") as file:
    pickle.dump(trimmed_data, file)
```
